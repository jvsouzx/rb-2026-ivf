```python
import numpy as np
import plotly.graph_objects as go

dataset = np.random.randint(1, 100, size=(200, 3))

fig = go.Figure(data=[
    go.Scatter3d(
        x=dataset[:, 0],
        y=dataset[:, 1],
        z=dataset[:, 2],
        mode="markers",
        marker=dict(size=4, color="#1f77b4", opacity=0.8),
        name="Dataset",
    )
])

fig.update_layout(
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis",
    ),
    margin=dict(l=0, r=0, b=0, t=20),
)

fig.show()
```

```python
def euclideanDistance(a: np.ndarray, b: np.ndarray):
    return np.sum((a - b) ** 2)


def kmeans(*, k: int, points: np.ndarray):
    centroids = np.random.randint(0, 100, size=(k, points.shape[1]))
    converged = False

    while not converged:
        clusters = [[] for _ in range(k)]

        for i in range(len(points)):
            point = points[i]
            closest_index = 0
            min_distance = euclideanDistance(point, centroids[0])
            for j in range(k):
                distance = euclideanDistance(point, centroids[j])
                if distance < min_distance:
                    min_distance = distance
                    closest_index = j
            clusters[closest_index].append(point)

        new_centroids = np.array([
            np.mean(cluster, axis=0) if len(cluster) > 0 else centroids[i]
            for i, cluster in enumerate(clusters)
        ])

        converged = np.allclose(centroids, new_centroids)
        centroids = new_centroids
        
    return clusters, centroids
```

```python
clusters, centroids = kmeans(k=3, points=dataset)
```

```python
cluster_colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd",
    "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf",
]

fig = go.Figure()

for i, cluster in enumerate(clusters):
    cluster = np.array(cluster)

    if len(cluster) == 0:
        continue

    fig.add_trace(
        go.Scatter3d(
            x=cluster[:, 0],
            y=cluster[:, 1],
            z=cluster[:, 2],
            mode="markers",
            marker=dict(
                size=4,
                color=cluster_colors[i % len(cluster_colors)],
                opacity=0.8,
            ),
            name=f"Cluster {i}",
        )
    )

fig.add_trace(
    go.Scatter3d(
        x=centroids[:, 0],
        y=centroids[:, 1],
        z=centroids[:, 2],
        mode="markers",
        marker=dict(
            size=7,
            color="black",
            line=dict(color="white", width=2),
        ),
        name="Centroids",
    )
)

fig.update_layout(
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis",
    ),
    legend=dict(itemsizing="constant"),
    margin=dict(l=0, r=0, b=0, t=20),
)

fig.show()
```
